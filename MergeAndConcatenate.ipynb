```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temperature" : ['32','45','30'],
    "humidity" : ['80','60','78']
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city":["new york","chicago","orlando"],
    "temperature" : ['21','14','35'],
    "humidity" : ['68','65','75']
})
us_weather
```

```python
df_weather = pd.concat([us_weather,india_weather],keys = ['us', 'india'])
df_weather
```

```python
temp_df = pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temperature" : ['32','45','30']
},index = [0,1,2])
temp_df
```

```python
windspeed_df = pd.DataFrame({
    "city":["delhi","mumbai"],
    "temperature" : ['45','32']
},index = [1,0])
windspeed_df
```

```python
pd.concat([temp_df,windspeed_df],axis = 1)
```

```python
df1 = pd.DataFrame({
    "city":["new york","chicago","orlando","baltimore"],
    "temperature" : ['21','14','35','38'],
})
df1
```

```python
df2 = pd.DataFrame({
    "city":["new york","chicago","san dieago"],
    "humidity" : ['68','65','75']
})
df2
```

```python
pd.merge(df1,df2,on = 'city')
```

```python
pd.merge(df1,df2,on = 'city',how = 'left')
```

```python
pd.merge(df1,df2,on = 'city',how = 'right')
```

```python
pd.merge(df1,df2,on = 'city',how = 'outer')
```
