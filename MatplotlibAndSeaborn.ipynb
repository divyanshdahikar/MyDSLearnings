```python
import pandas as pd 
```

```python
df_sales = pd.read_excel("linechart.xlsx")
df_sales
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize = (13,4))
plt.plot(df_sales["Quarter"], df_sales["Fridge"], color = 'blue',label = "Fridge")
plt.plot(df_sales["Quarter"], df_sales["Dishwasher"],color = 'green', label = "Dishwasher")
plt.plot(df_sales["Quarter"], df_sales["Washing Machine"],color = 'red',label = "Washing Machine")
plt.title("Product Sales(Mln $)")
plt.ylabel("Revenue(Mln $)")
plt.xlabel("Financial Quarter")
plt.legend()
plt.show()
```

```python
df_sales
```

```python
total_sales = df_sales[["Fridge", "Dishwasher", "Washing Machine"]].sum()
```

```python
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales, labels = total_sales.index, autopct = "%1.2f%%", explode = (0.1,0.1,0.1), startangle = 145, shadow = True)
plt.show()
```
